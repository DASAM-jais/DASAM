```python
#汇总数据
def get_all_subdirs(directory):
    # 获取目录下的所有文件和文件夹
    items = os.listdir(directory)
    # 过滤出所有文件夹
    subdirs = [item for item in items if os.path.isdir(os.path.join(directory, item))]
    return subdirs

def list_csv_files(path):
    csv_files = [f for f in os.listdir(path) if f.endswith('.csv')]
    return csv_files

# 定义一个函数来处理每个元素
def process_element(elem):
    if elem is None:
        return 0
    # 处理字符串列表
    try:
        elem = ast.literal_eval(elem)
        elem=np.array(elem)
    except (ValueError, SyntaxError):
        pass
    return elem

# 定义一个函数来计算列表中每个位置的平均值
def mean_of_list_elements(column):
    return np.mean(np.array(column), axis=0)
```

```python
# 指定要列出子文件夹的文件夹路径
folder_path = r'C:\Users\11054\PycharmProjects\pythonProject0\在线评论\数据\实验数据\基本实验3-扩充'
# 获取所有子文件夹
subfolders = get_all_subdirs(folder_path)
initial=0
for subfolder in subfolders:
    print(1)
    rps=subfolder.split('-')
    rp1=rps[2].replace('回应偏好','')
    rp2=rps[3].replace('回应偏好','')
    rp2=rp2.replace('.csv','')
    csv_files = list_csv_files(folder_path+'\\'+subfolder)#获取所有csv文件
    count=0
    max_count=1000
    for csv_file in csv_files:
        csv_file_path=folder_path+'\\'+subfolder+'\\'+csv_file
        csv_file=csv_file.replace('指标结果', "")
        csv_file=csv_file.replace('.csv', "")
        if initial==0:
            base_df=pd.read_csv(csv_file_path)
            base_df=base_df.applymap(process_element)
            base_df['回应偏好1']=float(rp1)
            base_df['回应偏好2']=float(rp2)
            base_df['random_seed']=count+1
        else:
            df=pd.read_csv(csv_file_path)
            df=df.applymap(process_element)
            df['回应偏好1']=float(rp1)
            df['回应偏好2']=float(rp2)
            df['random_seed']=count+1
            base_df=pd.concat([base_df,df])
        count+=1
        initial+=1
        if count%max_count==0:
            print(max_count)
            break
        
base_df.to_csv(r'C:\Users\11054\.spyder-py3\my_project\在线评论\在线评论-2024-9-14\数据-2024-9-14\data_all_扩充.csv', index=False, encoding='utf-8-sig')
```
